# Scatter Plot

A scatter plot is a way to visualize how multi-dimensional data are
distributed across certain values.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate quasi-linear data
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
coeff = 1.0 # TRY increase and decrese the coeef, to 0.5 to 2, to 5 ?
y = 2 * x + np.random.randn(100) * coeff  # Linear relation with some noise,

# Create a scatter plot
plt.scatter(x, y, color='blue')

# Add titles and labels
plt.title('Quasi-Linear Relationship Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the scatter plot
plt.show()
```

# Pie Chart

A pie chart is a way to visualize the static composition (aka, distribution) of a variable (or single group).


```python
import matplotlib.pyplot as plt

# Data to plot

labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [30, 25, 20, 205]  # Percentage values

# TRY: add more data, add Pears, 450

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)

# Add a title
plt.title('Fruit Distribution')

# Equal aspect ratio ensures the pie is drawn as a circle.
plt.axis('equal')

# Display the pie chart
plt.show()
```

# Stacked area graph
A stacked area graph is a way to visualize the composition of a group as it changes over time (or some other quantitative variable). This shows the relationship of a categorical variable (AgeGroup) to a quantitative variable (year).


```python
import matplotlib.pyplot as plt
import numpy as np

# Simulating cumulative data over time
x = np.arange(2000, 2021)  # Years from 2000 to 2020
y1 = np.random.randint(1, 5, size=len(x)).cumsum()  # Cumulative trend for series 1
y2 = np.random.randint(1, 4, size=len(x)).cumsum()  # Cumulative trend for series 2
y3 = np.random.randint(1, 3, size=len(x)).cumsum()  # Cumulative trend for series 3

# Stack the data

# Add titles and labels
plt.title('Cumulative Trends Over Time (2000-2020)')
plt.xlabel('Year')
plt.ylabel('Cumulative Value')


plt.stackplot(x, y1, y2, y3, labels=['Trend 1', 'Trend 2', 'Trend 3'], colors=['#FF9999', '#66B2FF', '#99FF99']) #UNCOMMENT
# Display the plot
plt.show()#UNCOMMENT
```

# Iris data set

The Iris flower dataset is one of the most well-known datasets in data science and machine learning. It was introduced by the British biologist and statistician Ronald A. Fisher in 1936 as part of his work on discriminant analysis.

## Overview of the Iris dataset:

Features (Columns): The dataset consists of four features:

* Sepal Length (in cm)
* Sepal Width (in cm)
* Petal Length (in cm)
* Petal Width (in cm)

Target (Label): There are three classes (species) of the Iris flower:

* Iris setosa
* Iris versicolor
* Iris virginica

Samples: The dataset contains 150 records, with 50 samples from each of the three species of Iris.

```python
# First, let's import the necessary libraries and load the Iris dataset to perform the exploratory data analysis (EDA).
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris_data = load_iris()
iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)


# 1. Pairplot for an overview of relationships
sns.pairplot(iris_df, hue='species', markers=["o", "s", "D"])
plt.suptitle('Pairplot of Iris Data Set', y=1.02)
plt.show()
```

```python
# 2. Distribution plots (histograms) for each feature
iris_df.iloc[:, :-1].hist(bins=20, figsize=(12, 8)) #TRY, more and less bins
plt.suptitle('Histogram of Features')
plt.show()
```

# Plotting 3D data with some tricks

```python
from matplotlib.lines import Line2D

# Prepare the data for 2D bubble plot
x = iris_df['sepal length (cm)']
y = iris_df['sepal width (cm)']
colors = iris_df['species'].map({'setosa': 'red', 'versicolor': 'blue', 'virginica': 'green'})
bubble_size = iris_df['petal width (cm)']  # Encoding sepal length as bubble size

# Create 2D bubble plot
plt.figure(figsize=(10, 7))
plt.scatter(x, y, s=bubble_size*150, c=colors, alpha=0.6, cmap='viridis')

# Add labels and title
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('2D Bubble Plot of Iris Dataset (Bubble Size: Petal Length)')

legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Setosa', markerfacecolor='red', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Versicolor', markerfacecolor='blue', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Virginica', markerfacecolor='green', markersize=10)
]
plt.legend(handles=legend_elements, title="Species")

# Show plot
plt.show()

 # Analyze !
```
